```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Assuming the data is in a CSV file named "cac40_data.csv"
df = pd.read_csv('CAC40.csv')

# Convert the 'Date' column to a datetime object
df['Date'] = pd.to_datetime(df['Date'])
```

```python
start_date = '2019-01-01'
end_date = '2023-01-01'

df_filtered = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
```

```python
# Calculate the mean, standard deviation, minimum, maximum, and median
mean_price = df_filtered['Adj Close'].mean()
std_price = df_filtered['Adj Close'].std()
min_price = df_filtered['Adj Close'].min()
max_price = df_filtered['Adj Close'].max()
median_price = df_filtered['Adj Close'].median()

# Print the results
print('Descriptive Statistics for CAC40 (2019-2023)')
print('------------------------------------------')
print(f"Mean Price: {mean_price:.2f}")
print(f"Standard Deviation: {std_price:.2f}")
print(f"Minimum Price: {min_price:.2f}")
print(f"Maximum Price: {max_price:.2f}")
print(f"Median Price: {median_price:.2f}")
```

```python
# Plot the CAC40 index over time
plt.figure(figsize=(10, 6))
plt.plot(df_filtered['Date'], df_filtered['Adj Close'])
plt.title('CAC40 Index (2019-2023)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
